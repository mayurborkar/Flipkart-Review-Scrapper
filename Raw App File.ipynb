```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + 'iphone7' # Website Name + Product Name Without Space
flipkart_url
```

```python
uClient = uReq(flipkart_url) #Is Try To Open Given Url With The Help Of URL Open
uClient
```

```python
flipkartPage = uClient.read() # Trying To Read The Operation
flipkartPage
```

```python
uClient.close() # To Closing URl Client After Reading The Information
```

```python
flipkart_html = bs(flipkartPage, "html.parser") # Calling Beautiful Soup So That We Can Segaregate The Data.
flipkart_html
```

```python
bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"}) # From That We Select Specific Product from Class & Division
bigboxes
```
